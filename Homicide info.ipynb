```python


data.plot('Year', color='red') 
data.plot('Victim Count', color='blue')

# Add the axis labels
plt.xlabel('Year')
plt.ylabel('Victim Count')

# Set the x-axis range
plt.xlim(1990, 2014)

# Set the y-axis range
plt.ylim(0, 1000)

# Add a title and display the plot
plt.title('Victim Count')
plt.show()
```
